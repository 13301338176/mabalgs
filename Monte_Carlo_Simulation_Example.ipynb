```python
import matplotlib.pyplot as plt
from mab import simulator
```

# Experimentation 1

```python
means = {0: [0.9, 0.1, 0.1, 0.1, 0.1]}
number_of_pulls = 3000
number_of_simulation = 1000

mcs_ths = simulator.MonteCarloSimulator()
results_ths = mcs_ths.run("ths", means, number_of_simulation, number_of_pulls)

mcs_ucb1 = simulator.MonteCarloSimulator()
results_ucb1 = mcs_ucb1.run("ucb1", means, number_of_simulation, number_of_pulls)

mcs_ucbtuned = simulator.MonteCarloSimulator()
results_ucbtuned = mcs_ucbtuned.run("tuned", means, number_of_simulation, number_of_pulls)
```

```python
plt.ylim(0.0, 1.0)
plt.title('Performance of Thompson Sampling')
plt.xlabel('Time')
plt.ylabel('Probability of select an arm')
plt.plot(range(results_ths[1]), results_ths[2][:,0], label='0')
plt.plot(range(results_ths[1]), results_ths[2][:,1], label='1')
plt.plot(range(results_ths[1]), results_ths[2][:,2], label='2')
plt.plot(range(results_ths[1]), results_ths[2][:,3], label='3')
plt.plot(range(results_ths[1]), results_ths[2][:,4], label='4')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.ylim(0.0, 1.0)
plt.title('Performance of UCB1')
plt.xlabel('Time')
plt.ylabel('Probability of select an arm')
plt.plot(range(results_ucb1[1]), results_ucb1[2][:,0], label='0')
plt.plot(range(results_ucb1[1]), results_ucb1[2][:,1], label='1')
plt.plot(range(results_ucb1[1]), results_ucb1[2][:,2], label='2')
plt.plot(range(results_ucb1[1]), results_ucb1[2][:,3], label='3')
plt.plot(range(results_ucb1[1]), results_ucb1[2][:,4], label='4')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.ylim(0.0, 1.0)
plt.title('Performance of UCB Tuned')
plt.xlabel('Time')
plt.ylabel('Probability of select an arm')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[2][:,0], label='0')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[2][:,1], label='1')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[2][:,2], label='2')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[2][:,3], label='3')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[2][:,4], label='4')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.title('Cumulative rewards of UCB1 vs Thompson Sampling vs UCB Tuned')
plt.xlabel('Time')
plt.ylabel('Rewards')
plt.plot(range(results_ucb1[1]), results_ucb1[3], label='UCB1')
plt.plot(range(results_ucb1[1]), results_ths[3], label='THS')
plt.plot(range(results_ucbtuned[1]), results_ucbtuned[3], label='UCB Tuned')
plt.legend()
plt.grid(True)

plt.show()
```

```python
print('Cumulative Rewards for UCB1: {}'.format(results_ucb1[3][-1]))
print('Cumulative Rewards for THS: {}'.format(results_ths[3][-1]))
print('Cumulative Rewards for UCB Tuned: {}'.format(results_ucbtuned[3][-1]))
```
